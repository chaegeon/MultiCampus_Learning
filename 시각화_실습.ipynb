```python
# _ 로 구분되어 있으니 split('_') 으로
names = []
for name in tmp.columns:
  if name == '행정구역':
    # print(name)
    names.append(name)
    continue
  # print( pd.to_datetime(name.split('_')[0], format='%Y년%m월') )
  names.append(pd.to_datetime(name.split('_')[0], format='%Y년%m월'))
```

```python
tmp.columns=names
tmp.head(2)
```

```python
tmp[pd.Timestamp('2008-01-01 00:00:00')].str.replace(',', '').astype(np.int)
```

```python
def conv(x):
  if x.name == '행정구역': return x
  return x.str.replace(',', '').astype(np.int)

tmp = tmp.apply( conv )
tmp.head(2)
```
